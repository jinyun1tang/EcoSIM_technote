## General description
EcoSIM is land biogeochemical model designed to simulate plant-soil-microbial interactions in natural and managed ecosystems under varying climate forcing. It adopts the mechanistically-based approach by employing physical rules to the maximum extent in describing transport, uptake, and chemical transformations of mass and energy. It uses the average representative approach to represent the plant and microbial populations. For plants, EcoSIM considers the interactions between biogeocheistry and biogeophysics in the context of explicitly tracking of each plant's morphological and phenolgoical development. This allows the characterization of plants in terms of measurable trait parameters that control light caputuring, photosynthesis and respiration, canopy and root phenolgoical and morphological development, drought and heat tolerance, nutrient acquisition, and geometrically-thermodynamically-constrained organ-specific biometric allocation and elemental stoichoimetry (C:N:P). For soil microbes, it represents 11 functional groups that conduct biogeochemistry involing organic matter, acetate, methane, reduced and oxidized nitrogen species, and oxygen in five oraganic matter complexes such as fine litter, coarse woody litter, manure, POM and humus. In the soil, plant roots and microbes compete for oxygen and mineral nutrients in the presence of modulations by aqueous geochemistry and multiphase transport (gaseous, aqueous, and precipitation-dissolution). 

Because EcoSIM focuses on mechanisms, it enables the many response parameters to emerge, a methodology that is differ significant from many other models. For instance, in biological growth, EcoSIM abandons the idea of fixed carbon use efficiency and nutrient use efficiency, rather it allows such response parameters to emerge from the interactions of more basic processes, e.g. storage, cost of assimilative uptake and maintenance. Other similar emergent parameters include temperature sensitivity of hetetrotrohpic respiration, plant respiration, and the response of photosynthesis to vapor pressure deficit.       

# EcoSIM uniqness

As a land biogeochemical model that explicitly couples microbial community dynamics with plant and soil biogeochemistry, EcoSIM distinguishes itself from other existing land biogeochemical models through the following major aspects:

* **Explicit representation of microbial processes**: EcoSIM explicitly represents microbes that carry out carbon decomposition, nitrogen and phosphorus cycling, and methane cycling. It also couples microbial biochemistry with aqueous geochemistry and surface chemistry. For each soil layer in the model, it has 13 microbial functional groups in five groups of organic matter complexes. It also includes nitrogen fixation in both the canopy and soil.

* **Detailed representation of plant morphology and phenology**: EcoSIM considers detailed plant morphology and geometry for seeds, fruits, primary and secondary roots, branches, and leaves. It also represents the entire lifecycle of plants, from seed planting to seed harvest. For typical crops, such as corn or maize, the represented phenology can be compared directly to the system used for field monitoring.

* **First-principles based formulation**: EcoSIM makes the maximum use of first-principles to formulate its model processes. This means that many relationships that are used as parameterizations in other models, such as photosynthetic Vmax and respiration Q10 in ELM, can emerge as special cases of model responses.

* **Mechanistic representation of ecosystem management**: EcoSIM's mechanistic representation allows for comprehensive treatment of ecosystem management practices, including fire, insect attack, grazing, crop rotation, diverse fertilizer application, N2O inhibition, tillage, irrigation, and ecosystem warming.
Explicit trait coupling between plants and microbes: EcoSIM explicitly couples plants and microbes through diverse biogeochemical and biogeophysical traits, which enables the simulation of ecosystem performance in the presence of concurrent competition for light, water, and nutrients.

* **Coupling of soil organic matter dynamics with soil physical properties**: EcoSIM explicitly couples soil organic matter dynamics with soil physical properties, such as heat capacity, soil porosity, and water conductivity. This enables a more complete coupling between ecosystem biogeochemistry and biogeophysics.


* **Scalability**: EcoSIM can simulate soil incubation, greenhouse experiments, and gridded regions using the same model parameterization.

## Forcing data

**Soil volume**

For each soil layer, 1 m$^3$ is seprated into rock volume, soil mass volume, macropore volume and micropore volume. After subtracting the rock volume, the total volume is separated into macropore, micropore and solid materials holding these pores. Total porosity is sum of macropores and micropores. The micropore soil volume refers to the volume of the soil fractions occupied by micropores.

Soil bulk density refers to the density of soil that excludes rock.

EcoSIM output relative soil saturation by dividing the water-filled pore volume (normalized by the total soil volume, excluding rock) with the soil pore volume, where the latter is computed with bulk density and particle density (as a function of soil organic matter).

Soil porosity is defined as volume of pore space in a unit volume of soil. The unit is m3/m3.

**Soil bulk density**

Soil bulk density is defined as the dry weight of soil per unit volume of soil. It is typically expressed in grams per cubic centimeter (g/cm¬≥). This measurement includes both the volume of soil particles and the volume of pores among soil particles.

Bulk density considers the combined volume of solids and pore space, whereas particle density only considers the mineral solids1. It is an indicator of soil compaction and can be calculated by dividing the oven-dry weight of soil by its volume at field moisture capacity or another specified moisture content.

For example, in an ideal soil where half is solids and half is pore space, the bulk density would be around 1.33 g/cm¬≥. However, bulk density can vary depending on soil texture, organic matter content, and management practices, typically ranging from 1.0 to 1.8 g/cm¬≥ for mineral soils.

Macropores (pores larger than 75 ¬µm) are part of the total pore space in soil and are therefore included in the volume used to calculate bulk density.

High macroporosity generally corresponds to lower bulk density because macropores increase the total volume without adding mass


**Relationship between rocks and bulk density in measurements:**

1. Inclusion of rocks:

Generally, when measuring soil bulk density, rocks and coarse fragments (particles >2 mm in size) are included in the measurement.

2. Challenges with rocky soils:

Soils with substantial amounts of coarse fragments present special challenges for bulk density measurements.
Inserting the core or probe may be difficult or impossible due to coarse fragments.

3. Effect on bulk density:
Coarse fragments are denser than soil, so their presence inside a sample will increase the apparent bulk density of the sample.

4. Measurement methods:
The core method, which is commonly used for bulk density measurements, includes rocks in the sample volume.
Some other methods, like excavation methods, may allow for separation and measurement of rock volume.

5. Considerations:
In some cases, researchers may choose to report bulk density on a "rock-free" basis, but this is not the standard approach.
When rocks are present, it's important to note their presence and potentially their volume fraction in the soil.
Importance of consistency:
Whatever method is chosen (including or excluding rocks), it should be applied consistently and reported clearly to ensure comparability of results.

# Tillage mixing

Tillage is defined by two parameters: tillage type and tillage depth. Tillage type is translated into fraction being mixed.

These two parameters are used to implement the following algorithm. 

1. Accumualte the total materials being mixed, meanwhile identify (1) the last layer being tilled, the fraction of each layer being tilled and not tilled.
2. Mix by distributing proportionally the cumulated materail to all tilled layers, using the equation:
   
   $V_{new}=f_{z,mixed}(V_{old}*(1-f_{mix})+TV*f_z)+(1-f_{z,mixed})*V_{old}$

where $f_{mix}$ is the level of mixing, $f_{z,mixed}$ is the thickness of the layer being mixed, and $f_{mix}$ is the mixing intensity.

Solutes are distributed in both macropores and micropores and solid materails are assumed to form the pore structure. 


For macropore and micropore exchange, it assumes that tillage does not change the distribution of macropores and tillage scrabs a fraction of materials from macropores into micropores. Materials from the surface litter layer are scrabed into macpore pores of the mixed layers.

Information about xylem and phloem distribution in grass/tree stems/stalks:

Scattered vascular bundles: In grass stems (culms), the xylem is distributed in scattered vascular bundles throughout the stem cross-section, rather than being arranged in concentric rings as in trees.

Bundle arrangement: These vascular bundles, which contain both xylem and phloem, are dispersed across the stem rather than being organized in a ring near the edge.

Parallel orientation: The vascular bundles in grass stems generally run parallel to the length of the stem.

Fixed number: Unlike trees, grasses do not have secondary growth, so they don't add new layers of xylem each year. The number of vascular bundles is fixed, but they elongate as the plant grows.

Structural support: While the xylem in grass stems provides some structural support, it's not as significant as in woody plants. Grasses rely more on other tissues, like sclerenchyma, for mechanical strength.

*Maturity group* 
Maturity group is an important parameter controlling phenology. For instance, in soybean, it classifies cultivars by the timing of their key developmental stages‚Äîespecially flowering (R1), seed filling (R5), and physiological maturity (R8)‚Äîin response to environmental cues like photoperiod and temperature. In EcoSIM, it affects the carbon allocation to different organs, particularly  to leaves, and petiole.

Plants with greater maturity group matures later.

Maturity group (MG) is a classification system ranging from very early (MG 000) to very late (MG X) based mostly on the length of the crop‚Äôs life cycle from emergence to maturity under specific photoperiod conditions

Each MG has typical durations for vegetative (emergence to flowering) and reproductive (flowering to maturity) phases. Earlier MGs complete these stages faster (suited for northern latitudes/short seasons), while later MGs have extended periods (suited for southern latitudes/longer seasons)

Maturity group for maize is often defined based on degree days.


Maturity groups are used to optimize planting and breeding decisions for several crop plants beyond soybean and maize. Some other crop plants commonly classified with maturity groups for optimization include:

### 1. **Cotton**
- Maturity groups classify cultivars based on time to physiological maturity to match regional growing seasons.
- Helps in selecting varieties for early or late harvest, optimizing yield and fiber quality.

### 2. **Wheat**
- Hard red winter and spring wheat varieties are grouped by maturity, e.g., early, intermediate, and late maturing types.
- This classification helps adapt wheat to planting dates and climatic zones to optimize grain fill and yield.

### 3. **Rice**
- Rice varieties are grouped into maturity classes (early, medium, late) based on days to maturity.
- This aids in crop scheduling across multiple plantings per year and adapting to regional climate.

### 4. **Sorghum**
- Sorghum hybrids are classified by maturity length based on heat units or days to maturity.
- Allows selection of cultivars suitable for varying climates and cropping systems.

### 5. **Sunflower**
- Maturity groups classify hybrids based on their growth duration and flowering time.
- Important for aligning planting to regional growing conditions and harvest timing.

The key phenological stages of annual grasses typically include:
1. Germination: Seeds germinate and emerge from the soil, usually in fall or early spring depending on the species and environmental conditions.
2. Vegetative growth: The grass produces leaves and tillers, expanding its biomass. This stage is characterized by rapid growth and leaf production.
3. Stem elongation: The stems begin to elongate and become jointed, with nodes and internodes becoming visible.
4. Boot stage: The developing seed head is enclosed within the sheath of the flag leaf but not yet visible.
5. Heading: The seed head emerges from the leaf sheath. This marks the transition from vegetative to reproductive growth.
6. Flowering: Anthers emerge and pollen is released. This stage is crucial for seed production.
7. Seed development: Seeds form and mature within the seed heads.
8. Senescence: The plant begins to dry out and turn brown as seeds reach maturity.
9. Seed dispersal: Mature seeds are dispersed, often by wind or animal movement.
10. Death: The annual plant completes its lifecycle and dies, leaving behind seeds for the next generation.

# Hypocotyl height

Hypocotyl height refers to the length of the stem-like structure in a seedling that is located below the cotyledons (seed leaves) and above the radicle (root). The hypocotyl plays a crucial role in seedling development and can vary in length depending on several factors:
Growth conditions: Hypocotyl length can differ significantly based on light exposure and the presence of growth regulators. For example:
In dark conditions, hypocotyls tend to be much longer (17.15 mm at stage IV) compared to light conditions (1.84 mm without GA and 2.93 mm with GA at stage IV)

1. The addition of gibberellic acid (GA) can increase hypocotyl length in light conditions1.
Plant species: Different plant species may have varying hypocotyl lengths.
Germination type: In epigeal germination, the hypocotyl elongates and pushes the cotyledons above the soil surface. In hypogeal germination, the hypocotyl remains short5.

Developmental stage: Hypocotyl length increases as the seedling develops. For instance, in Arabidopsis, the hypocotyl length progresses from 0.31 mm at stage I to 1.84 mm at stage IV under light conditions without GA

# Canopy energy balance (UptakesMod.F90)

The first order closure of canopy heat storage is

$$
\frac{d}{dt}\left(C_{H,Z}T_C\right)=H_{in}+C_{a}(T_{air}-T_C) \tag{1}
$$

where $H_{in}$ is net heat input due to radiation ($R_{net}$), precipitation/dew and evaportranspiration (and the associated convective heat). $C_a$ is the sensible heat conductance between atmosphere and canopy. $C_{H,Z}$ is the canopy heat capacity, including contributions from canopy organic matter, and water held in biomass and on canopy.

When equation (1) is discretized, it becomes

$$
C_{H,Z,t+1}T_{C,t+1}=C_{H,Z,t}T_{C,t}+H_{in}\Delta t+C_{a}(T_{air}-T_{C,t+1})\Delta t \tag{2}
$$

From which, $T_{C,t+1}$ can be solved as

$$
T_{C,t+1}=\frac{C_{H,Z,t}T_{C,t}+H_{in}\Delta t+C_{a}T_{air}\Delta t}{C_{H,Z,t+1}+C_{a}T_{air}\Delta t} \tag{3}
$$

# Root solute/nutrient uptake

Root nutrient uptake is done considering the balance between MM-kinetics based uptake and solute supplied from advection and diffusion. The mathematical formulation is as follows 

$$qC+D(C-C_0)=V_{max}\frac{C_0-C_m}{K+C_0-C_m}=Q$$

where on the left of the first "=" is solute supply through transpiration-induced advection and concentration gradient induced diffusion; the term on the right of the first "=" is MM-kinetics, considering compensation point $C_m$, and the uptake flux is designated by $Q$. $C_0$ is the solute concentration at the uptake site, $C$ is the solute concentration in soil. 

From above, we can first solve for $C_0$ as a function of $Q$ and the MM uptake as

$$C_0=\frac{QK}{V_{max}-Q}+C_m$$

which then is used to formulate the quadratic equation as

$$(q+D)C-D(\frac{QK}{V_{max}-Q}+C_m)=Q$$

Define $X=(q+D)C$, $Y=DC_m$, then we obtain

$$X-Y-D\frac{QK}{V_{max}-Q}=Q$$

which can be re-written as

$$Q^2-(V_{max}+X-Y+DK)Q+(X-Y)V_{max}=0$$

Then by introducing $\alpha=-(V_{max}+X-Y+DK)<0$, and $\beta=(X-Y)V_{max}>0$, we have

$$Q^2+\alpha Q+\beta=0$$

The uptake is then found as  

$$Q=\frac{1}{2}\left(-\alpha-\sqrt{\alpha^2-4\beta}\right)$$

Key parameters of the above model are 
  1. compensation point: $C_m$,   
  1. maximum uptake rate $V_{max}$, and   
  1. solute diffusivity $D$.


#  The effective air temperature and vapor pressure for sensible and latent heat exchange with respect to ground surfaces (litter, snow and bare soil).

$$

$$

$$

$$

$$

$$

Here, $r_{b,h}$ and $r_{b,w}$ are boundary layer resistance for sensible and latent heat transport respectively. $A$ is area, $L$ is lantent heat of evaporation, and $A$ is total surface area. 

Moreover, assuming there is no heat stored in air, there are relationships 

$$

Hydraulic gradient $= (h_‚ÇÇ - h_‚ÇÅ) / L$

$h_‚ÇÅ$ is the hydraulic head at the upstream point

$L$ is the distance between the two points (scalar)

# Freeze-thaw calculation 
1. Compute the apparent temperature  $T_{app}$ based on incoming energy from water and energy flow and heat source/sink.
2. Compute freezing temperature $T_F$ based on the Clausis-Clapeyron equation.
3. Compute the energy available for phase change by bring down the material temerpature to freezing temperature $T_F$, which is $C_p(T_{app}-T_F)$.
4. Do phase change using the avaialble phase change energy and compute the energy used for phase change.
5. If all avilable phase change energy is used, the the final temeprature if $T_F$.
6. If there is more available phase energy than used for phase change, do cooling or warming according to the updated heat capacity.

Specifically, do macropore and micropore freeze-thaw separately according to

   In the macropore freeze-thaw calculation, only ice and water are involved.
   In the micropore freeze-thaw calculation, soil solid material, water and ice are involved.
   
   Use the excessive enthalpy to do freeze water or thaw ice.

   If the excessive thalpy is consumed, then the final temperature is at $T_F$; else the update heat capacity, and use the excessive enthalpy to heat or cool the material.

   The enthalpy conservation is then: $C_{p,total,0}*T_0=C_{p,total,new}*T_1+H_{phase}$

# Canopy/stomatal conductance

The canopy conductance (for transpiratin) is weighted betweent the cuticular resistance, and the minimum resistance based on a weighting function with leaf turgor pressure as driving variable. Specifically,

$$r_{can}=r_{can,min}+(r_{cuti}-r_{can,min})exp(\alpha_{shp}\Psi_{turg})$$
where $\alpha_{shp} < 0$ with unit [1/MPa], and $\Psi_{turg}>0$ with unit MPa. Typical values of turgor pressure is about 1 MPa. $r_{cuti}$ is the cuticular resistance, the maximum resistance.

# Surface-atmosphere gas exchange

There are two-ways of surface-atmosphere gas exchange: (1) Litter-atmosphere exchange, and (2) soil-atmosphere exchange. The exchange is represented as **diffusion flux** between equilbrium aqueous concentration and the aqueous concentration in litter and soil, with a proper accounting of resistance due to tortuosity, snow and atmosphere.

Since the gas concentration in the litter layer is set to atmospheric concentration, a dissolution flux also occurs.

Other fluxes include tracers coming from **wet deposition** through irrigation and precipitation.

There are other fluxes through **ebullition and plant-aided transport**.

# How plants die
Under normal conditions, e.g. excluding cutting, fire, plants dies from carbon starvation. In a drought, the plant first lose structural leaf biomass, then with remobilization, it increases the osmolyte concentration, which lowers the osmotic pressure, and then increases turogr pressure. In real world, higher turgor pressure can lead to physical blow up. However, in EcoSIM, the continuous loss of leave reduces photosynthesis and thereby starvation and death of the plant.

In general, higher plants are more vulnerable to drought attack.

# Ebullition
**Hydrostatic pressure**
It is recognized that the soil depth is much smaller in magnitude that the thickness of atmosphere, the hydrostatic pressure in unsaturated soil thus equals to the atmospheric pressure. For any depth below the water table, the hydrostic pressure is the atmospheric pressure plus the pressure exerted by the relative height of water $Z_{wt}-Z_{soil}$.

**The gas pressure**
Based on the Henry's law, the equilbrium gas molar concentration $C_g$ and its corresponding dissolved mass $m_{s,w}$ has the relationship

$$
m_{s,w}=M_w C_g\alpha \theta_w
$$

where $\theta_w$ is the volume of liquid water, and $M_w$ is the molecular weight.

The corresponding gas pressure is then

$$
P_g=C_gRT=\frac{m_{s,w}RT}{M_w\alpha\theta_w}
$$

Ebullition is modeled as an vaporization process. When the total gas pressure from all different volatiles exceeds the atmospheric pressure (+ standing water), aqueous concentrations are reduced and added to the gaseous tracers, which then undergoes convection and diffusion. The bulling flux at the surface is added to efflux. In soil depths below water tabble, hydrostatic pressure is increased due to saturated thickness above. 

**Total hydrostatic pressure at any soil depth ($P_z$)**

$P_z$ equals to the atmospheric pressure and the pressue induced by standing water from above. If depths above is unsaturated, then $P_z$ equals to atmospheric pressure, given distance from soil surface to $z$ is small.

# Micro-site gas uptake

Oxygen and CH$_4$ uptake by a microsite (only in micropores) is solved by iterations.

In each iteration, both gaseous and aqueous phases are updated, where uptake is through aqueous phase, dissolution occurs between gaseous and aqueous phases. The gaseous and aqueous phases are also subject to external fluxes (advection from other neighbor grids for an inside soil grid, or from atmosphere diffusion for litter layer), as diagnosed from previous model time step.  

For CH4 and likely H2, CH4 uptake flux may occasionally be greater than CH4 concentrations, even though the algorithm has been carefully to avoid this. One consquence is the soil CH4 (and likely H2 as well) gas concentrations are zero, which should be interepted as beyond the detection limit. This seems to be an issue related to the formulation of CH4 oxidation as demand driven, where energy is first computed as a function concentrations (in both gaseous and aqueous phases), but then then uptake for biomass is scaled with the available catabolic energy. The supply driven formulation will be able to resolve this issue.

The uptake is formulated as the balance between diffusion and MM-based uptake, such that

$$Q=D(C_s-C_0)=V_{max}\frac{C_0}{K+C_0}$$

By sovling for $C_0$ as $$C_0=C_s-\frac{Q}{D}$$, 
we have $$Q=V_{max}\frac{C_s-\frac{Q}{D}}{K+C_s-\frac{Q}{D}}=V_{max}\frac{C_sD-Q}{KD+C_sD-Q}$$, which can be re-written as

$$Q^2-(V_{max}+KD+C_sD)Q+V_{max}C_sD=0$$

Define $\alpha=-(V_{max}+KD+C_sD)<0$ and $\beta=V_{max}C_sD$, we have

$$Q=\frac{1}{2}(-\alpha-\sqrt{\alpha^2-4\beta})$$


Gas advection is done based on expansion/shrink due to water flow, assuming it is incompressible.

# Tile draiange

Title draiange occurs only for soil layers that are above tile pipes. The water table of a soil column $D_1$ is determined as

$$D_1=\left(d_s-L_i\frac{\theta_i}{\theta_{i,s}}\right)$$

where $d_s$ [m], $L_i$ [m], $\theta_i$ [m$^3$ m$^{-3}$] and $\theta_{i,s}$ [m$^3$ m$^{-3}$] are depth to the top of uppermost saturated soil layer, thickness of the lowest unsaturated  soil layer, volumetric soil water content of the lowest unsaturated soil layer, and saturated volumetric soil water content of the lowest unsaturated soil layer.

The tile flow is modeled using the Darcy's formulation

$$v=K\frac{\Delta D}{d} $$

where $v$ [m h$^{-1}$], $K$ [m h$^{-1}$], and $\Delta D$ [m] are flow velocity, saturated hydraulic conductance and the pressure drop over a distance $d$ [m].

# Plant water uptake
Plant water uptake follows the conceptual model by Reid and Huck (1990). Specifically, it assumes each soil layer is connected to canopy independently via the roots in each soil layer. The gradient between canopy water potential, and the depth-corrected soil water potential drives the water uptake from a given soil layer. Soil water is assumed to first to overcome the soil resistance to reach the root/mycorrhizae surface, then overcomes radial resistance and finally axial resistance to reach the canopy. 

However, as total water upatke is constrained by energy supply at the surface, water is allowed to move between different soil layers with plant serving as a router, assuming perfect mixing. For a soil receives water, it is not able to know where the water come from. 

How to compute radial resistance? Qualitatively, greater root surface area will reduce radial resistance. 

How to compute axial resistance? Qualitatively, longer root will increase axial resistance. 

Greater number of roots will lower both axial and radial resistance.

For parameterization, radial resistance is input as for per $m^2$ of root surface, while axial resistance is for per m of root length. Because root surface area is a function of root length and root radius, root thickness modulates the radial resistance and 

It was noted that the model by Reid and Huck (1990) was devised for herbaceous crops. In order to apply for trees, axial resistance is further divided into two components due to primary roots and secondary/fine roots. The primary roots then add canopy height/size control on axial resistance for water transport.

Its governing equation can be written as
$$
\frac{\partial V_r\theta_r}{\partial t}=\frac{\partial }{\partial z}\left(K_{axial}\frac{\partial \psi_r}{\partial z}\right)+\frac{\psi_w-\psi_0}{R_{soil}}
$$

$$
\frac{\psi_w-\psi_0}{R_{soil}}=\frac{\psi_0-\psi_r}{R_{radial}}
$$

Here $\psi_w$ is soil water pressure (with depth correction), $\psi_0$ is the matric pressure at root surface, and $\psi_r$ is root pressure. $V_r$ is root volume, $\theta_r$ is root moisture content. $R_{soil}$ is the soil resistance from soil to root surface, $R_{radial}$ is radial resistance for water to go from root surface into xylem. $K_{axial}$ is axial condutance for water tarnsport via xylem.

Assuming biomass water content $\theta_r$ changes negiligiblly, we then have

$$
\theta_r\frac{\partial V_r}{\partial t}=\frac{\partial }{\partial z}\left(K_{axial}\frac{\partial \psi_r}{\partial z}\right)+\frac{\psi_w-\psi_0}{R_{soil}}
$$


Because the above equation is linear, it is equivalent to 
$$
\theta_r\left(\frac{\partial V_r}{\partial t}\right)_l=\frac{\partial }{\partial z}\left(K_{axial}\frac{\partial \psi_r}{\partial z}\right)+\left(\frac{\psi_w-\psi_0}{R_{soil}+R_{radial}}\right)_l=\frac{\partial }{\partial z}\left(K_{axial}\frac{\partial \psi_r}{\partial z}\right)+U_l
$$
and $U=\sum_{l}U_l$ is the total water uptake. Here $\theta_r\left(\frac{\partial V_r}{\partial t}\right)_l$ is the contribution to water for biological growth due to uptake from layer $l$.

The above equation can be recast into the integral form as

$$
\theta_r\int_{z_l}^{Z_C}{\left(\frac{\partial V_r}{\partial t}\right)_l}dz=\int_{z_l}^{Z_C}{d\left(K_{axial}\frac{\partial \psi_r}{\partial z}\right)}+U_l\Delta_l
$$
where $U_l\Delta_l$ is water uptake from layer $l$, whose thickness is $\Delta_l$.

Moreover
$$
\int_{z_l}^{Z_C}{d\left(K_{axial}\frac{\partial \psi_r}{\partial z}\right)}=T_l=\frac{K_{stalk}}{H_C}\left(\psi_C-\psi_b\right)=\frac{K_{l2b}}{Z_l}\left(\psi_b-\psi_{r,l,0}\right)=\frac{\psi_{r,l,0}-\psi_{r,l}}{R_{axial,l}}
$$

$$
U=\sum_l{\left\lbrack T_l+\theta_r\left(\frac{\partial V}{\partial t}\right)\Delta_l+\theta_r\left(\int_{z_{0}}^{Z_c}\frac{\partial V}{\partial t}dz\right)_l\right\rbrack}+\theta_r\int_{z_{0}}^{Z_c}\frac{\partial V}{\partial t}dz=T+\theta_r\sum_l{\frac{\partial V_l}{\partial t}\Delta_l}+\theta_r\int_{z_{0}}^{Z_c}{\frac{\partial V}{\partial t}}dz
$$

# Radial and axial resistance 
By approximating xylem as tubes, application of Poiseuille's law leads to an estiamted axial resistance about the order of  $c\times 10^9 \mathrm{~MPa~s~m}^{-4}$, where $c$ is varies between 1 to 100. This is equivalent to $c_x\times 10^5 \mathrm{~MPa~h~m}^{-4}$. Based on the measurement by Frensch and Steudle (1990) for maize roots, the radial resistance is of the similar order, i.e. $c_R\times 10^5 \mathrm{~MPa~h~m}^{-4}$, although the scalar $c_R$ may be very likely smaller than $c_X$. The axial resistance is a function of xylem radius, and number of xylem per root, two varaibles that are related to the root radius. Physically, greater xylem radius and more xylem tubes there are, lower is the axial resistance. The radial resistance is related to the effective root surface area (covered by aqua porins) for water uptake. Azaizeh and Steudle (1991) indicate that under salty conditions, the magnitude can be of $10^7 \mathrm{~MPa~s~m}^{-4}$. The total root conductance is on the order of $10^{-7} \mathrm{m s}^{-1} \mathrm{MPa}^{-1}$.

# Metabolic coupling of microbes
Microbes carries out catabolic reactions that produces Gibbs free energy, which is stored as ATPs (with an efficiency $\lambda$) that are later used to drive anabolic activity. This process is quantified using the respiration efficiency $\epsilon$ via

$$\epsilon=\frac{1}{1+\lambda \Delta G_c/\Delta G_a}$$

All coupling goes into steps:
1. synthesis of CH$_2$O (from CO$_2$/CH$_4$ for autotrophs), invovling $G_c$ and $G_a$.
1. gross respiration of CH$_2$ into CO$_2$/CH$_4$, which is used for maintenance and growth, involving $G_r$ and $G_a$.

The microbial metabolism is computed as follows: 
1. Compute substrate uptake $U_{c}$, 
2. Partition it into catabolic reaction $R_c$, and gross respiration $R_p=R_c*Y_c*e_r$, where $Y_c$ is (intermediate) C yield and $e_r$ is respiration ratio of using the intermediate C for growth. 
   $Y_c$ is based on the Gibbs free energy of catabolic reaction and the anabolic reaction that converts carbon substrate into intermediate carbon biomass. The intermediate carbon is oxidized to produce gross respiration with a (respiration) yield ratio $e_r$. 
4. Apply oxygen limitation to $R_p$ and $R_c$ for aerobic organisms. 
5. Determine growth respiration $R_g$ via $R_g=R_p-R_m$, where $R_m$ is maintenance respiration.
6. Obtain the actual substrate uptake $U_s=min(R_m,R_p)+R_g/e_r+R_c$.

It is noted that for the biomass growth rate $R_g*(1/e_r-1)=(R_p-R_m)*(1/e_r-1)$, some electron donor/acceptors are assumed to be provided from inside the cell. In short, $R_c*Y_c$ is partitioned between maintenance respiration ($R_m/e_r$), growth respiration and biomass growth.


# Methanogens

**Hydrogenotrophic methanogen** uses $\mathrm{H}_2$ as electron donor, and $\mathrm{CO}_2$ as electron acceptor, and $\mathrm{CO}_2$ as carbon source for biomass. It uses the catbolic reaction

$$\Delta G_c:\mathrm{CO}_2+4\mathrm{H}_2 \rightarrow \mathrm{CH}_4+2\mathrm{H}_2\mathrm{O}$$

The typical value of $\Delta G_c$ is -131kJ molC$^{-1}$, and $G_c=G_{c,0}+RTlog\left(\frac{[H_2]}{[H_2]_0}\right)^4$.

Its anabolic reaction is 

$$\Delta G_a:2\mathrm{H}_2(aq)+\mathrm{CO}_2(aq)\rightarrow \mathrm{CH}_2\mathrm{O}+\mathrm{H}_2\mathrm{O}$$

$\Delta G_a$ is about 38.9 kJ molC$^{-1}$.

The respiration reaction is 
$$G_r:\mathrm{CH}_2\mathrm{O}+2\mathrm{H}_2 (aq)\rightarrow \mathrm{CH}_4(aq) + \mathrm{H}_2\mathrm{O}$$

$G_r$ is about ‚àí162 kJ (molC)$^{-1}$.

From above, H$_2$ is consumed in both catabolic and anabolic reactions, with $R_c$ used in catabolic reaction and $R_a$ used in anabolic reaction.  

In step 1. The yield is $\frac{G_a}{G_a+G_c}$, in step 2, the yield is $\frac{G_a}{G_r+G_a}$.

The actual implementation differs from the above paragradim, instead, all respiration is assumed to follow

$$\mathrm{CO}_2(aq)+4\mathrm{H}_2 (aq)\rightarrow \mathrm{CH}_4 (aq)+ 2\mathrm{H}_2\mathrm{O}$$
Therefore, $R_p$ is directly related to substrate uptake. Respiration coefficient $e_r$ is computed based on the Gibbs free energy $G_c$ and $G_a$. And $\mathrm{CO}_2$ is converted directly into biomass.

# Ammonia oxidizer
It uses the (mole-based) catabolic reaction 
$$\Delta G_c:\mathrm{NH}_3+1.5 \mathrm{O}_2\rightarrow \mathrm{NO}_2^-+ \mathrm{H}^+ + \mathrm{H}_2\mathrm{O}$$
$\Delta G_c$ is about ‚àí260 to ‚àí275 kJ molN$^{-1}$.


Respiration (mole-based)
$$\Delta G_r:\mathrm{CH}_2\mathrm{O}+\mathrm{O}_2\rightarrow \mathrm{CO}_2+\mathrm{H}_2\mathrm{O}$$
$\Delta G_r$ is about -478 kJ molC$^{-1}$.

The (mole-based) anabolic reaction
$$\Delta G_a: 1.5\mathrm{CO}_2+\mathrm{NH}_3 +\mathrm{OH}^-\rightarrow 1.5\mathrm{CH}_2\mathrm{O}+\mathrm{NO}_2^-+0.5\mathrm{H_2}\mathrm{O}$$
which in the mass-based form
$$\mathrm{CO}_2+0.778\mathrm{NH}_3 +\mathrm{OH}^-\rightarrow \mathrm{CH}_2\mathrm{O}+0.778\mathrm{NO}_2^-+0.5\mathrm{H_2}\mathrm{O}$$
$G_a$ is about 3500~4685 kJ molC$^{-1}$. The ratio $G_c/G_a$ is about 0.08. So 0.08/(1+0.08)=0.0741 of NH$_3$ is used for producing CH$_2$O.

$\mathrm{NH}_3$ is oxidized by $\mathrm{NO}_2^-$ into N$_2$O via

$$G_{r,NO2}: \mathrm{NH}_3(aq)+2\mathrm{NO}_2^- \rightarrow 1.5\mathrm{N}_2\mathrm{O}(aq)+2\mathrm{OH}^-+\mathrm{H}_2\mathrm{O}$$

The value of $G_{r,NO2}$ is about -249.8kJ (mol NH3)$^{-1}$. This gives $G_{c,NO2}/G_a$ about 0.07. So 0.07/(1+0.07)=0.0654 of NH$_3$ is used for producing CH2O.


# Nitrite oxidizer
It uses the (mole-based) catabolic reaction 
$$G_c: \mathrm{NO}_2^-+0.5\mathrm{O}_2 (aq) \rightarrow \mathrm{NO}_3^-$$
where $G_c$ is about ‚àí5.9 kJ (molN)$^{-1}$.

The anabolic reaction used is
$$G_a: \mathrm{CO}_2+2\mathrm{NO}_2^-+\mathrm{H}_2\mathrm{O}\rightarrow 2\mathrm{NO}_3^-+\mathrm{CH}_2\mathrm{O}$$
$G_a$ is about 384 kJ (molC)$^{-1}$.

$$G_r:\mathrm{CH}_2\mathrm{O} + \mathrm{O}_2 \rightarrow \mathrm{CO}_2 + \mathrm{H}_2\mathrm{O}$$.

As above, $\mathrm{NO}_2^-$ is consumed both in the catabolic reaction and the synthesis of $\mathrm{CH}_2\mathrm{O}$, but $\mathrm{CO}_2$ is used for all three processes. In other words, $U_{NO2}$ is partitioned for catabilic reaction and the synthesis of $\mathrm{CH}_2\mathrm{O}$, the part of $\mathrm{CH}_2\mathrm{O}$ is $U_{NO2}Y_c$, which is used for gross respiration and growth. Because respiration is achieved through the oxidation of $\mathrm{CH}_2\mathrm{O}$, the overall reaction is still $\mathrm{NO}_2^-+0.5\mathrm{O}_2\rightarrow \mathrm{NO}_3^-$.

# Aerobic methane oxidizer
$$G_a:\mathrm{CH}_4(aq)+\mathrm{O}_2(aq)\rightarrow \mathrm{CH}_2\mathrm{O}+\mathrm{H}_2\mathrm{O}$$

$$G_c:\mathrm{CH}_2\mathrm{O}(aq)+\mathrm{O}_2(aq) \rightarrow \mathrm{CO}_2(aq)+\mathrm{H}_2\mathrm{O}$$

$$G_r: \mathrm{CH}_2\mathrm{O} + \mathrm{O}_2 \rightarrow \mathrm{CO}_2+\mathrm{H}_2\mathrm{O}$$
$G_c$ is about ‚àí809 kJ (molC)$^{-1}$. 
$G_a$ is about 550 kJ (molC)$^{-1}$.

One mole of ATP carries 30.5 kJ/mol. The serine pathway takes 1.5 ATP per HCHO. The RuMP pathway takes one ATP for 3HCHO. In generally, Type-I with RuMP pathway grows faster than type-II using the serine pathway.

For all aerobic autotrophs, the inorganic C, i.e. $\mathrm{CH}_4$ or $\mathrm{CO}_2$, is first converted into CH$_2$O, which is then oxidized into CO$_2$ to compute gross respiration.
